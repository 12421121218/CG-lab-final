```python
import matplotlib.pyplot as plt

# Chosen triangle coordinates
A = (2, 1)
B = (5, 1)
C = (4, 4)

triangle = [A, B, C]

# Reflection functions
def reflect_x(points):
    return [(x, -y) for (x, y) in points]

def reflect_y(points):
    return [(-x, y) for (x, y) in points]

# Apply reflections
triangle_x = reflect_x(triangle)
triangle_y = reflect_y(triangle)

# Function to get plotting coordinates (close the triangle)
def get_plot_coords(points):
    xs = [p[0] for p in points] + [points[0][0]]
    ys = [p[1] for p in points] + [points[0][1]]
    return xs, ys

orig_x, orig_y = get_plot_coords(triangle)
rx_x, rx_y = get_plot_coords(triangle_x)
ry_x, ry_y = get_plot_coords(triangle_y)

# Plotting
plt.figure(figsize=(7,7))
plt.plot(orig_x, orig_y, 'b-o', label='Original Triangle', linewidth=2)
plt.plot(rx_x, rx_y, 'r-o', label='Reflected X-axis', linewidth=2)
plt.plot(ry_x, ry_y, 'g-o', label='Reflected Y-axis', linewidth=2)

# Annotate points
for i, (x, y) in enumerate(triangle):
    plt.text(x + 0.1, y + 0.1, f"P{i}({x},{y})", color='blue', fontsize=9)
for i, (x, y) in enumerate(triangle_x):
    plt.text(x + 0.1, y + 0.1, f"P{i}_X({x},{y})", color='red', fontsize=9)
for i, (x, y) in enumerate(triangle_y):
    plt.text(x + 0.1, y + 0.1, f"P{i}_Y({x},{y})", color='green', fontsize=9)

# Axes and grid
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')
plt.legend()
plt.title("2D Reflection of Triangle along X and Y axes")
plt.show()

# Print coordinates
print("Reflection along X-axis:")
for i, p in enumerate(triangle_x):
    print(f"P{i}_X = {p}")

print("\nReflection along Y-axis:")
for i, p in enumerate(triangle_y):
    print(f"P{i}_Y = {p}")
```
