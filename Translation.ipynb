```python
import matplotlib.pyplot as plt
import math

# Function to calculate distance between two points
def distance(p1, p2):
    return math.sqrt((p2[0]-p1[0])**2 + (p2[1]-p1[1])**2)

# Function to check if four points form a square
def is_square(points):
    A, B, C, D = points
    d = sorted([distance(A,B), distance(B,C), distance(C,D), distance(D,A), distance(A,C), distance(B,D)])
    # For square: 4 equal sides, 2 equal diagonals (sides < diagonals)
    return math.isclose(d[0], d[1]) and math.isclose(d[0], d[2]) and math.isclose(d[0], d[3]) and math.isclose(d[4], d[5])

# Input square points (order: A,B,C,D)
A = (0,0)
B = (3,0)
C = (3,3)
D = (0,3)

points = [A, B, C, D]

if not is_square(points):
    print("Warning: The points do not form a perfect square!")

# Translation distances
tx = 1
ty = 1

# Calculate translated points
translated_points = [(x+tx, y+ty) for (x,y) in points]

# Prepare for plotting (close the square)
original_x = [p[0] for p in points] + [points[0][0]]
original_y = [p[1] for p in points] + [points[0][1]]

translated_x = [p[0] for p in translated_points] + [translated_points[0][0]]
translated_y = [p[1] for p in translated_points] + [translated_points[0][1]]

# Plotting
plt.figure(figsize=(6,6))
plt.plot(original_x, original_y, 'b-o', label='Original Square', linewidth=2)
plt.plot(translated_x, translated_y, 'r-o', label='Translated Square', linewidth=2)

# Annotate original points
for i, (x,y) in enumerate(points):
    plt.text(x + 0.1, y + 0.1, f"P{i}({x},{y})", color='blue', fontsize=9)

# Annotate translated points
for i, (x,y) in enumerate(translated_points):
    plt.text(x + 0.1, y + 0.1, f"P{i}'({x},{y})", color='red', fontsize=9)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')
plt.legend()
plt.title("2D Translation of a Square")
plt.show()

# Print new coordinates
print("New coordinates after translation:")
for i, p in enumerate(translated_points):
    print(f"P{i}' = {p}")
```
