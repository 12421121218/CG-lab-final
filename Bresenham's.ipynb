```python
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2):
    points = []

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1

    if dx > dy:
        err = dx // 2
        y = y1
        for x in range(x1, x2 + sx, sx):
            points.append((x, y))
            err -= dy
            if err < 0:
                y += sy
                err += dx
    else:
        err = dy // 2
        x = x1
        for y in range(y1, y2 + sy, sy):
            points.append((x, y))
            err -= dx
            if err < 0:
                x += sx
                err += dy

    return points

# Example: line from (2, 2) to (7, 8)
x1, y1 = 2, 2
x2, y2 = 7, 8

points = bresenham(x1, y1, x2, y2)

print("Calculated Points (Integer):")
for p in points:
    print(p)

x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

plt.figure(figsize=(6,6))
plt.plot(x_vals, y_vals, marker='o', color='blue')
for i, (x, y) in enumerate(points):
    plt.text(x + 0.1, y + 0.1, f"P{i}({x},{y})", fontsize=9)
plt.title("Bresenham's Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
