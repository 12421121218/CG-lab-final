```python
import matplotlib.pyplot as plt
import math

A = (0, 0)
B = (1, 0)
C = (1, 1)

triangle = [A, B, C]

def rotate_point(point, angle_deg):
    angle_rad = math.radians(angle_deg)
    x, y = point
    x_new = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    y_new = x * math.sin(angle_rad) + y * math.cos(angle_rad)
    return (round(x_new, 2), round(y_new, 2))

angle = 90  # anticlockwise

rotated_triangle = [rotate_point(p, angle) for p in triangle]

original_x = [p[0] for p in triangle] + [triangle[0][0]]
original_y = [p[1] for p in triangle] + [triangle[0][1]]

rotated_x = [p[0] for p in rotated_triangle] + [rotated_triangle[0][0]]
rotated_y = [p[1] for p in rotated_triangle] + [rotated_triangle[0][1]]

# Plotting
plt.figure(figsize=(6,6))
plt.plot(original_x, original_y, 'b-o', label='Original Triangle', linewidth=2)
plt.plot(rotated_x, rotated_y, 'r-o', label='Rotated Triangle', linewidth=2)

# Annotate points
for i, (x,y) in enumerate(triangle):
    plt.text(x + 0.05, y + 0.05, f"P{i}({x},{y})", color='blue', fontsize=9)
for i, (x,y) in enumerate(rotated_triangle):
    plt.text(x + 0.05, y + 0.05, f"P{i}'({x},{y})", color='red', fontsize=9)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')
plt.legend()
plt.title("2D Rotation of Triangle (90 Anticlockwise)")
plt.show()


print("New coordinates after 90 anticlockwise rotation:")
for i, p in enumerate(rotated_triangle):
    print(f"P{i}' = {p}")
```
