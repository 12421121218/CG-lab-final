```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    for _ in range(steps + 1):
        px = round(x)
        py = round(y)
        points.append((px, py))  # Keep as integers
        x += x_inc
        y += y_inc
    return points

x1, y1 = 1, 1
x2, y2 = 6, 6
points = dda(x1, y1, x2, y2)

print("Calculated Points (Rounded to Integer):")
for p in points:
    print(p)

plt.figure(figsize=(6,6))
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

plt.plot(x_vals, y_vals, marker='o', color='blue')

for i, (x, y) in enumerate(points):
    label = f"P{i}({x},{y})"
    plt.text(x + 0.1, y + 0.1, label, fontsize=9)

plt.title("DDA Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
