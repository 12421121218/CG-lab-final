```python
import matplotlib.pyplot as plt

def bresenham_circle(xc, yc, r):
    points = []
    x = 0
    y = r
    d = 3 - 2 * r

    while x <= y:
        # 8 symmetric points
        points.append((xc + x, yc + y))
        points.append((xc - x, yc + y))
        points.append((xc + x, yc - y))
        points.append((xc - x, yc - y))
        points.append((xc + y, yc + x))
        points.append((xc - y, yc + x))
        points.append((xc + y, yc - x))
        points.append((xc - y, yc - x))

        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        x += 1

    return points

# Circle parameters
xc, yc = 0, 0
radius = 6
points = bresenham_circle(xc, yc, radius)

print("Calculated Points of Circle:")
for p in points:
    print(p)

# Plotting
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

plt.figure(figsize=(6,6))
plt.scatter(x_vals, y_vals, color='green')

# Annotate each point
for i, (x, y) in enumerate(points):
    plt.text(x + 0.2, y + 0.2, f"({x},{y})", fontsize=8)

plt.title("Bresenham's Circle Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
