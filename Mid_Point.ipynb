```python
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    points = []

    dx = x2 - x1
    dy = y2 - y1

    # Check if slope is steep
    steep = abs(dy) > abs(dx)

    if steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2
        dx, dy = x2 - x1, y2 - y1

    # Swap points if necessary
    if x1 > x2:
        x1, x2 = x2, x1
        y1, y2 = y2, y1

    dx = x2 - x1
    dy = y2 - y1
    yi = 1 if dy >= 0 else -1
    dy = abs(dy)

    d = 2 * dy - dx
    y = y1

    for x in range(x1, x2 + 1):
        if steep:
            points.append((y, x))
        else:
            points.append((x, y))
        if d > 0:
            y += yi
            d -= 2 * dx
        d += 2 * dy

    return points

# Example: line from (3, 6) to (6, 11)
x1, y1 = 3, 6
x2, y2 = 6, 11
points = midpoint_line(x1, y1, x2, y2)

print("Calculated Points (Integer):")
for p in points:
    print(p)

# Plotting
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

plt.plot(x_vals, y_vals, marker='o', color='green')
for i, (x, y) in enumerate(points):
    plt.text(x + 0.1, y + 0.1, f"P{i}({x},{y})", fontsize=9)
plt.title("Midpoint Line Drawing Algorithm")
plt.grid(True)
plt.axis('equal')
plt.show()
```
